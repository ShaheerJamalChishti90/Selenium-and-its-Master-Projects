# <b>This is the scraping of Coin Market Website</b> 
## <b>using Selenium</b>


```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
```

```python
options = Options()
driver = webdriver.Chrome(options=options)
driver.get("https://coinmarketcap.com/")
```

```python
tables = driver.find_element(By.CLASS_NAME, "cmc-table")
tables
```

```python
all_rows = tables.find_elements(By.TAG_NAME, "tr")
all_rows
```

#### <b style="color: yellow;">at this point, ye mujhe complete data nahi de raha tha, uske liye mujhe apne driver walay browser ko complete scroll karna para jis ki wajha se page ka data save hogaya, ab jab page ka data save hogaya tou mujhe wo rows wala block of code re-run karna para takay ab mujhe complete rows de jitni page scroll karny ki wajha se save hogyii thi, jab usne mujhe complete 100 rows de di tou maine neechay wala block of code run kia jab jaa kar mujhe complete page ka data(100 rows ka data) mila</b>

```python
for row in all_rows[1:]:
    table = row.find_elements(By.TAG_NAME, "td")
    rank = table[1].text
    name = table[2].text
    price = table[3].text
    hour_change = table[4].text
    day_change = table[5].text
    seven_days_change = table[6].text
    market_cap = table[7].text
    volume24hour = table[8].text
    
    print(rank, name, price, hour_change, day_change, seven_days_change, market_cap, volume24hour)
```

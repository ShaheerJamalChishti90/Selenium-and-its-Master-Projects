# 1. Scraping with BS4


```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://coinmarketcap.com/"
response = requests.get(url).content
soup = BeautifulSoup(response, "lxml")
```

```python
table = soup.find("table", {"class" : "sc-db1da501-3 ccGPRR cmc-table"})
table
```

```python
all_trs = table.find_all("tr")
all_trs
```

### This is the 01 way to extract the data out

```python
for rows in all_trs:
    table_data = rows.find_all("td")
    for data in table_data:
        print(data.text)
    
    
    # print(table_data)
```

### This is the 02 way to extract the data out

```python
for rows in all_trs[1:]:
    table_data = rows.find_all("td")
    rank = table_data[1].text
    name = table_data[2].text
    price = table_data[3].text
    market_cap = table_data[4].text
    # volume_24h = table_data[5].text
    # circulating_supply = table_data[5].text
    # change_24h = table_data[6].text
    
    print(f"Rank: {rank}")
    print(f"Name: {name}")
    print(f"Price: {price}")
    print(f"Market Cap: {market_cap}")
    # print(f"Volume (24h): {volume_24h}")
    # print(f"Circulating Supply: {circulating_supply}")
    # print(f"Change (24h): {change_24h}")
    print("-" * 20)
```
